```python
x = reuters.paras(fileids=['training/9897'])
# printing original list 
print("The original list : " + str(x))
```

```python
def get_list_values(text, temp=[]):
    for item in text:
        if type(item) == list:
            temp = get_list_values(item, temp)

        else:
            temp.append(item)

    return temp

 
res = ' '.join(get_list_values(x))
res
```

```python
res = res.replace(" . ", ".")
```

```python
res = res.replace("pct","%")
res
```

```python
import spacy
from spacy import displacy
import en_core_web_sm

# Create an instance of the small pipeline and model from SpaCy
nlp = en_core_web_sm.load()

# Apply the complete NER pipeline from SpaCy (preprocessing + model prediction)
doc = nlp(res)

# Show the entities detected in a nice way
displacy.render(doc, jupyter=True, style='ent')
```

```python
import pandas as pd
import nltk
named_entities = []
for sentence in res:
    temp_entity_name = ''
    temp_named_entity = None
    sentence = nlp(res)
    for word in sentence:
        term = word.text 
        tag = word.ent_type_
        if tag:
            temp_entity_name = ' '.join([temp_entity_name, term]).strip()
            temp_named_entity = (temp_entity_name, tag)
        else:
            if temp_named_entity:
                named_entities.append(temp_named_entity)
                temp_entity_name = ''
                temp_named_entity = None

entity_frame = pd.DataFrame(named_entities, 
                            columns=['Entity Name', 'Entity Type'])

```
