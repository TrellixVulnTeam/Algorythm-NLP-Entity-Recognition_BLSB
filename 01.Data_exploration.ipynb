### Perform pre-processing on the document from reuters dataset.

```python
def get_list_values(text, temp=[]):
    for item in text:
        if type(item) == list:
            temp = get_list_values(item, temp)

        else:
            temp.append(item)

    return temp

 
result = ' '.join(get_list_values(x))
result
```

```python
resultult = result.replace(" . ", ".")
result = result.replace("pct","%")
result
```

```python
import spacy
from spacy import displacy
import en_core_web_sm

# Create an instance of the small pipeline and model from SpaCy
nlp = en_core_web_sm.load()

# Apply the complete NER pipeline from SpaCy (preprocessing + model prediction)
doc = nlp(result)

# Show the entities detected in a nice way
displacy.render(doc, jupyter=True, style='ent')
```

## NER using Spacy

#### Extracting the entities and labels 

```python
import pandas as pd
import nltk

doc = nlp(result)

entities = []
labels = []

for ent in doc.ents:
    entities.append(ent)
    labels.append(ent.label_)
    
df = pd.DataFrame({'Entities':entities,'Labels':labels})

df
```

```python
top_entities = (df.groupby(by=['Entities', 'Labels'])
                           .size()
                           .sort_values(ascending=False)
                           .reset_index().rename(columns={0 : 'Frequency'}))
```

```python
named_entities = []
for sentence in result:
    temp_entity_name = ''
    temp_named_entity = None
    sentence = nlp(result)
    for word in sentence:
        term = word.text 
        tag = word.ent_type_
        if tag:
            temp_entity_name = ' '.join([temp_entity_name, term]).strip()
            temp_named_entity = (temp_entity_name, tag)
        else:
            if temp_named_entity:
                named_entities.append(temp_named_entity)
                temp_entity_name = ''
                temp_named_entity = None

entity_frame = pd.DataFrame(named_entities, 
                            columns=['Entity Name', 'Entity Type'])

```

```python
# get the top named entities
top_entities = (entity_frame.groupby(by=['Entity Name', 'Entity Type'])
                           .size()
                           .sort_values(ascending=False)
                           .reset_index().rename(columns={0 : 'Frequency'}))
top_entities.T.iloc[:,:17]
```
